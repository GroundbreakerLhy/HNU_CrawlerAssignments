##### **增加学生成绩信息**


使用循环判断语句，添加学生信息，包括学号、姓名、语文、数学、英语、物理、化学、生物、历史、地理、政治等。其中设置科目成绩范围为0~100

```python
import pandas as pd
# 增加学生成绩
def add_stu():
    while True:
        print_message("添加学生成绩！")
        # 添加学号
        stu_id_flag = True
        while stu_id_flag:
            stu_id = input("请输入学号：")
            if stu_id in ["exit","quit", 'q']:
                print_message("退出添加学生成绩！")
                break
            # 判断学号是否为空
            if stu_id == "":
                print_message("学号不能为空，请重新输入！")
                continue
            # 判断学号是否正确
            if len(stu_id) != 8:
                print_message("学号长度为8，请重新输入！")
                continue
            # 判断学号是否存在
            stus = read_csv()
            stu_ids = stus["学号"].values.tolist()
            if int(stu_id) in stu_ids:
                print_message("该学号已经存在，请重新输入！")
                continue
            else:
                stu_id_flag = False
                # 添加学生姓名
                stu_name_flag = True
                while stu_name_flag:
                    name = input("请输入学生姓名：")
                    if len(name) < 2 or len(name) > 10:
                        print_message("学生姓名长度为2-10，请重新输入！")
                        continue
                    else:
                        stu_name_flag = False
                        
                # 添加语文成绩
                stu_yuwen_flag = True
                while stu_yuwen_flag:
                    yuwen = input("请输入语文成绩：")
                    if int(yuwen) < 0 or int(yuwen) > 100:
                        print_message("成绩范围为0-100，请重新输入！")
                        continue
                    else:
                        stu_yuwen_flag = False
                    
                # 添加数学成绩
                stu_shuxue_flag = True
                while stu_shuxue_flag:
                    shuxue = input("请输入数学成绩：")
                    if int(shuxue) < 0 or int(shuxue) > 100:
                        print_message("成绩范围为0-100，请重新输入！")
                        continue
                    else:
                        stu_shuxue_flag = False

                # 添加英语成绩
                stu_yingyu_flag = True
                while stu_yingyu_flag:
                    yingyu = input("请输入英语成绩：")
                    if int(yingyu) < 0 or int(yingyu) > 100:
                        print_message("成绩范围为0-100，请重新输入！")
                        continue
                    else:
                        stu_yingyu_flag = False

                # 添加物理成绩
                stu_wuli_flag = True
                while stu_wuli_flag:
                    wuli = input("请输入物理成绩：")
                    if int(wuli) < 0 or int(wuli) > 100:
                        print_message("成绩范围为0-100，请重新输入！")
                        continue
                    else:
                        stu_wuli_flag = False

                # 添加化学成绩
                stu_huaxue_flag = True
                while stu_huaxue_flag:
                    huaxue = input("请输入化学成绩：")
                    if int(huaxue) < 0 or int(huaxue) > 100:
                        print_message("成绩范围为0-100，请重新输入！")
                        continue
                    else:
                        stu_huaxue_flag = False

                # 添加生物成绩
                stu_shengwu_flag = True
                while stu_shengwu_flag:
                    shengwu = input("请输入生物成绩：")
                    if int(shengwu) < 0 or int(shengwu) > 100:
                        print_message("成绩范围为0-100，请重新输入！")
                        continue
                    else:
                        stu_shengwu_flag = False

                # 添加历史成绩
                stu_lishi_flag = True
                while stu_lishi_flag:
                    lishi = input("请输入历史成绩：")
                    if int(lishi) < 0 or int(lishi) > 100:
                        print_message("成绩范围为0-100，请重新输入！")
                        continue
                    else:
                        stu_lishi_flag = False

                # 添加地理成绩
                stu_dili_flag = True
                while stu_dili_flag:
                    dili = input("请输入地理成绩：")
                    if int(dili) < 0 or int(dili) > 100:
                        print_message("成绩范围为0-100，请重新输入！")
                        continue
                    else:
                        stu_dili_flag = False

                # 添加政治成绩
                stu_zhengzhi_flag = True
                while stu_zhengzhi_flag:
                    zhengzhi = input("请输入政治成绩：")
                    if int(zhengzhi) < 0 or int(zhengzhi) > 100:
                        print_message("成绩范围为0-100，请重新输入！")
                        continue
                    else:
                        stu_zhengzhi_flag = False
                
                stu_dict = {"学号":stu_id,"姓名":name,"语文":yuwen,"数学":shuxue,"英语":yingyu,"物理":wuli,"化学":huaxue,"生物":shengwu,"历史":lishi,"地理":dili,"政治":zhengzhi}
                stu_df = pd.DataFrame(stu_dict, index=[0])
                new_stus = pd.concat([stus, stu_df], ignore_index=True, sort=False)
                new_stus.to_csv("stu.csv", index=False)
                print_message("添加学生成绩成功")
                display(stu_df)
        break

# add_stu()
```



##### **查询学生成绩信息**

通过学生姓名查询已有学生成绩。比如：

输入： 程和昶

输出：



```python
# 查询某个学生成绩
def read_someone():
    print_message("查询学生成绩！")
    stus = read_csv()
    display(stus)
    stu_names = stus["姓名"].values.tolist()
    while True:
        stu_name = input("请输入学生姓名：")
        if stu_name not in stu_names:
            print_message("该学生不存在，请重新输入！")
            continue
        else:
            stu = stus.loc[(stus['姓名'] == stu_name),:]
            print_message("【{}】该学生成绩如下：".format(stu_name))
            display(stu)
            break
# read_someone()
```


##### **修改学生成绩信息**

根据学生名字查询到学生的成绩，在已有参数列表中选择要修改的科目和成绩。比如原始成绩为：



输入：祝福你你

修改为：语文，99

输出：



```python
# 修改学生成绩

def update_someone():
    print_message("修改学生成绩！")
    stus = read_csv()
    display(stus)
    stu_names = stus["姓名"].values.tolist()
    while True:
        # 学生姓名输入逻辑
        stu_name_flag = True
        while stu_name_flag:
            stu_name = input("请输入待修改成绩学生姓名：")
            if stu_name not in stu_names:
                print_message("该学生不存在，请重新输入！")
                continue
            else:
                stu_name_flag = False
        # 学生课程输入逻辑
        stu_subject_flag = True
        while stu_subject_flag:
            subject = input("请输入【{}】要修改的科目：".format(stu_name))
            if subject not in ["语文","数学","英语","物理","化学","生物","历史","地理","政治"]:
                print_message("课程错误，请重新输入")
                continue
            else:
                stu_subject_flag = False
        # 学生成绩输入逻辑
        stu_grade_flag = True
        while stu_grade_flag:
            grade = input("请输入【{}】要修改的分数：".format(subject))
            if int(grade) not in range(0,100+1):
                print_message("成绩输入超出，请重新输入")
                continue
            else:
                stu_grade_flag = False
        # 获取输入课程index
        subject_index = list(stus.columns).index(subject)
        # 获取学生index
        stu_index = stus.loc[(stus['姓名'] == stu_name),:].index[0]
        # 修改对应index上的成绩（数值）
        stus.iloc[stu_index,subject_index] = int(grade)
        stus.to_csv("stu.csv", index = False)
        print_message("成绩修改完成")
        # 展示修改后的成绩
        new_stu = stus.loc[(stus['姓名'] == stu_name),:]
        display(new_stu)
        break

# update_someone()
```

#####  **删除学生成绩信息**

通过输入要删除的学生成绩信息的名字，可以选择是否将学生所有信息全部删除。比如删除前的学习成绩信息图为：



输入：张珊，y

输出：删除成功



```python
# 删除学生成绩

def delete_someone():
    print_message("删除学生成绩！")
    stus = read_csv()
    display(stus)
    stu_names = stus["姓名"].values.tolist()
    while True:
        # 学生姓名输入逻辑
        stu_name_flag = True
        while stu_name_flag:
            stu_name = input("请输入待删除成绩学生姓名：")
            if stu_name not in stu_names:
                print_message("该学生不存在，请重新输入！")
                continue
            else:
                stu_name_flag = False
        stu = stus.loc[(stus['姓名'] == stu_name),:]
        display(stu)
        delete_flag = input("确认是否删除学生【{}】成绩（y/n）".format(stu_name_flag))
        if delete_flag == "y":
            save_books = stus.drop(stus[stus['姓名']== stu_name].index)
            save_books.to_csv("stu.csv", index = False)
            display(save_books)
            print_message("【{}】成绩删除成功".format(stu_name))
            break
        else:
            print_message("退出删除")
            break

# delete_someone()
```

**学生总成绩柱状图**

```python
def grade_bar_sort(df):
    """学生总成绩柱状图"""
    # 获取学生的总成绩
    grade = df.iloc[:, 2:].sum(axis=1)
    # 获取对应成绩的学生姓名
    name = df.iloc[:, 1]
    # 数据合并
    zf_df = pd.concat([name, grade], axis=1)
    zf_df.columns = ['name', 'gross_score']
    
    # 根据学生总成绩进行降序排序
    sort_df = zf_df.sort_values(by=['gross_score'], ascending=False)
#     print(sort_df)
    width,height,size = 28, 12, 18
    # 默认画布大小
    plt.figure()                        
    plt.figure(figsize=(width,height))
    # 绘制柱形图
    plt.bar(range(0, len(sort_df)*6, 6), sort_df.iloc[:, 1], width=4)
    # 定义x轴的间隔，rotation：文字角度
    plt.xticks(list(range(0, 246, 6)), sort_df.iloc[:, 0], rotation=90, fontsize=size)
    # 柱子上的数字显示
    for a,b in zip(np.arange(0, len(sort_df.iloc[:, 0])*6, 6), sort_df.iloc[:, 1]):   
        plt.text(a,b,'%d'%b,ha='center',va='bottom',fontsize=size);
    plt.show()
```

```python
def grade_barh_sort(df):
    """学生总成绩水平柱状图"""
    # 获取学生总成绩
    grade = df.iloc[:, 2:].sum(axis=1)
    # 获取学生成绩对应姓名
    name = df.iloc[:, 1]
    # 数据合并
    zf_df = pd.concat([name, grade], axis=1)
    zf_df.columns = ['name', 'gross_score']
    # 根据学生总成绩进行降序排序
    sort_df = zf_df.sort_values(by=['gross_score'])
#     print(sort_df)
    width,height,size = 20, 24, 12
    # 默认画布大小
    plt.figure()                        
    plt.figure(figsize=(width,height))
    # 绘制水平柱状图
    plt.barh(range(0, len(sort_df)*3, 3), sort_df.iloc[:, 1], height=2)
    # 设置y轴的间隔
    plt.yticks(list(range(0, 123, 3)), sort_df.iloc[:, 0], fontsize=size)
    # 柱子上的数字显示
    for a,b in zip(np.arange(0, len(sort_df.iloc[:, 0])*3, 3), sort_df.iloc[:, 1]):   
         plt.text(b+10,a-0.5,'%d'%b,ha='center',va='bottom',fontsize=size);
    plt.show()
```

**学生各科成绩等级比例扇形图**

```python
def grade_pass_count(df):
    """学生各科成绩等级比例扇形图"""
    # 获取学生各科成绩
    subjects = df.iloc[:, 1:]
    width,height = 24, 18
    # 默认画布大小
    plt.figure()                        
    plt.figure(figsize=(width,height))
    # 定义学生成绩等级
    text = ["优秀","良好","中等","及格","不及格"]
    # 对每科学生成绩进行等级划分
    for i,name in enumerate(subjects.iloc[:, 1:].columns):
        # 等级条件，>=90为优秀，<90&>=80为良好，<80&>=70为中等，<70&>=60为及格，<60为不及格
        a = len(subjects[subjects[name]>=90])
        b = len(subjects[(subjects[name]<90) & (subjects[name]>=80)])
        c = len(subjects[(subjects[name]<80) & (subjects[name]>=70)])
        d = len(subjects[(subjects[name]<70) & (subjects[name]>=60)])
        e = len(subjects[subjects[name]<60])

#         print(name,a,b,c,d,e,text)
        data = [a,b,c,d,e]
        print(data, text)
        # 绘制饼图
        plt.pie(data,labels=text,autopct="%.2f%%")
        plt.subplot(331+i)
        plt.title(name)
    plt.show()
 
```

**各科学生成绩前三柱状图**

```python
def grade_max_bar(df):
    """各科学生成绩前三柱状图"""
    # 获取学生各科成绩
    subjects = df.iloc[:, 1:]
    # 获取各科的名字
    names = list(subjects.iloc[:, 1:].columns)
    datas = []
    # 获取各科成绩前三的数据列表
    for name in subjects.iloc[:, 1:].columns:
        max3 = subjects.sort_values(by=name,ascending=False)[['姓名', name]][:3]
        d = [x[0] for x in list(max3[[name]].values)]
        datas.append(d)
    # 转换成df结构
    data_df = pd.DataFrame(np.array(datas))
    width,height = 26, 18
    # 默认画布大小
    plt.figure()                        
    plt.figure(figsize=(width,height))
    
    x =list(range(0, len(names)*3, 3))
    total_width, n = 0.8, 2
    width = total_width / n

    # 绘制各科第一名成绩柱形图
    plt.bar(x, data_df[0], width=width, label='第一名',tick_label=names,fc = 'y')
    # 柱子上的数字显示
    for a,b in zip(np.arange(0, len(data_df[0])*3, 3), data_df[0]):   
         plt.text(a,b,'%d'%b,ha='center',va='bottom',fontsize=12);
    for i in range(len(x)):
        x[i] = x[i] + width
    # 绘制各科第一名成绩柱形图
    plt.bar(x,data_df[1], width=width, label='第二名',tick_label=names, fc = 'r')
    # 柱子上的数字显示
    for a,b in zip(np.arange(0, len(data_df[1])*3, 3), data_df[1]):   
        plt.text(a+width,b,'%d'%b,ha='center',va='bottom',fontsize=12);
    for i in range(len(x)):
        x[i] = x[i] + width
    # 绘制各科第一名成绩柱形图
    plt.bar(x,data_df[2], width=width, label='第三名',tick_label=names, fc = 'g')
    # 柱子上的数字显示
    for a,b in zip(np.arange(0, len(data_df[2])*3, 3), data_df[2]):   
        plt.text(a+width*2,b,'%d'%b,ha='center',va='bottom',fontsize=12);
    plt.legend()
    plt.show()
```

```python
def grade_max_bars(df):
    """各科学生成绩前三详细柱状图"""
    # 获取学生各科成绩
    subjects = df.iloc[:, 1:]
    # 获取各科名字
    infos = list(subjects.iloc[:, 1:].columns)
    names = []
    k = []
    n = 0
    # 获取各科学生成绩前三、对应学生姓名
    for name in subjects.iloc[:, 1:].columns:
        max3 = subjects.sort_values(by=name,ascending=False)[['姓名', name]][:3]
        d = [x[0] for x in max3[['姓名']].values]
#         print(max3[[name]].values[0].tolist())
        h = [x[0] for x in list(max3[[name]].values)]
        k.append(h)
        names.append(d)
        
        width,height = 20, 10
        # 默认画布大小
        plt.figure()                        
        plt.figure(figsize=(width,height))
        # 定义子图
        plt.subplot(331+n)
        # 绘制柱形图
        plt.bar(range(len(max3)), k[n], tick_label=names[n])
        # 显示柱形图数据
        for a,b in zip(np.arange(len(k[n])), k[n]):   
            plt.text(a,b,'%d'%b,ha='center',va='bottom',fontsize=12);
        n+=1
        # 显示每科标签名字
        plt.title(name)
        # 定义y轴范围
        plt.ylim(0,110)
    plt.show()
```

```python
def grade_avg_bar(df):
    """学生每科的平均成绩的柱状图"""
    subjects = df.iloc[:, 2:]
    names = list(subjects.columns)
    avg = subjects.mean()
    width, height, size = 20, 10, 12
    colors = ['green', 'blue', 'red', 'orange', 'purple', 'yellow', 'pink', 'cyan', 'magenta', 'brown']
    plt.figure()     
    plt.figure(figsize=(width, height))                  
    plt.grid()
    plt.bar(range(len(names)), avg, tick_label=names, color=colors[:len(names)])
    for a, b in zip(np.arange(len(avg)), avg):   
        plt.text(a, b, '%d' % b, ha='center', va='bottom', fontsize=size)
    plt.show()
```

```python
grade_avg_bar(df)
```

```python
def grade_avg_plot(df):
    """学生每科的平均成绩的折线图"""
    subjects = df.iloc[:, 2:]
    names = list(subjects.columns)
    avg = subjects.mean()
    width, height, size = 20, 10, 12
    plt.figure(figsize=(width, height))
    
    line_style = '-'
    line_color = 'blue'
    
    plt.plot(names, avg, linestyle=line_style, color=line_color)
    
    plt.xlabel('科目')
    plt.ylabel('平均成绩')
    plt.title('学生每科的平均成绩')
    
    # Add data labels
    for a, b in zip(np.arange(len(avg)), avg):   
        plt.text(a, b, '%d' % b, ha='center', va='bottom', fontsize=size)
    
    plt.show()
```

```python
grade_avg_plot(df)
```

```python
# 学生每科的平均成绩的可视化
def grade_avg_radar(df):
    """学生每科的平均成绩的雷达图"""
    subjects = df.iloc[:, 2:]
    names = list(subjects.columns)
    avg = subjects.mean()
    width,height,size = 20, 10, 12
    plt.figure()                        
    plt.figure(figsize=(width,height))
    plt.polar(np.linspace(0, 2*np.pi, len(names), endpoint=False), avg)
    plt.fill(np.linspace(0, 2*np.pi, len(names), endpoint=False), avg, 'b', alpha=0.3)
    plt.show()
```

```python
grade_avg_radar(df)
```
